A transform is orthogonal if their basis fuctions are orthogonal. In the discrete case, the basis functions can be generated through $\text{iT}(\delta(t+\tau))$, where iT stands for inverse transform ad $\tau$ is an index of the basis.

Two random variables $X$ and $Y$ are orthogonal if
\begin{equation}
  E(XY)=0.
\end{equation}

```python
# Forward and backward DCT

from scipy import fftpack

# Forward and backward DCT wrappers

def DCT(samples):
    return fftpack.dct(samples, norm="ortho")

def iDCT(coefficients):
    return fftpack.idct(coefficients, norm="ortho")
```
