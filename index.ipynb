A transform is orthogonal if their basis fuctions are orthogonal. In the discrete case, the basis functions can be generated through $\text{iT}(\delta(t+\tau))$, where iT stands for inverse transform ad $\tau$ is an index of the basis.

Two random variables $X$ and $Y$ are orthogonal if
\begin{equation}
  E(XY)=0.
\end{equation}

```python
# Forward and backward DCT

from scipy import fftpack

# Forward and backward DCT wrappers

def DCT(samples):
    return fftpack.dct(samples, norm="ortho")

def iDCT(coefficients):
    return fftpack.idct(coefficients, norm="ortho")
```

```python
N = 256
for x in range(0,N,10):
    coeffs = np.zeros(N)
    coeffs[x] = 1000.0
    basis_function_1 = iDCT(coeffs)
    for y in range(1,N,10):
        coeffs = np.zeros(N)
        coeffs[y] = 1000.0
        basis_function_2 = iDCT(coeffs)
        product = (basis_function_1 * basis_function_2).astype(np.int)
        expected_value = np.average(product)
        #basis_range = int(product.max() - product.min())
        #PDF, _x = np.histogram(product, bins=basis_range)
        #expected_value = expectation(product, PDF)
        print(expected_value)
```

Orthogonal transform generates uncorrelated coefficients. Therefore, the autocorrelation of the coefficients should be zero.

```python
def autocorr(x):
    result = np.correlate(x, x, mode='full')
    return result#[result.size//2:]
```

```python
N = 32
mu, sigma = 0, 0.1
signal = (255*np.random.normal(mu, sigma, size=N)).astype(int) # Gaussian
print(signal)
coeffs = DCT(signal)
print(coeffs)
print(autocorr(coeffs))
```

Another different test consists in measuring the quantization error in the transform and the signal domains. If both errors are the same, the transform is orthogonal.

```python
q_step = 10
N = 32
mu, sigma = 0, 0.1
#signal = (255*np.random.normal(mu, sigma, size=N)).astype(int) # Gaussian
signal = np.random.randint(255, size=N) # Uniform
print("signal =", signal)
coeffs = DCT(signal)
print("coefficients =", coeffs)
quantized_coeffs = (coeffs/q_step).astype(np.int)*q_step
print("quantized coefficients = ", quantized_coeffs)
quantization_error_coeffs_domain = coeffs - quantized_coeffs
print("quantization error in the transform domain = ", quantization_error_coeffs_domain, np.sum(quantization_error_coeffs_domain**2))
reconstructed_signal = iDCT(quantized_coeffs)
print("reconstructed signal =", reconstructed_signal)
quantization_error_signal_domain = signal - reconstructed_signal
print("quantization error in the signal domain = ", quantization_error_signal_domain, np.sum(quantization_error_signal_domain**2))
```
